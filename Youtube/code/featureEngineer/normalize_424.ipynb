```python
import pandas as pd
import numpy as np
from sklearn import preprocessing 
# combine with tag and title
df = pd.read_csv('addedFeatures.csv')

titles_and_tags = df[["title","tags"]]
titles_length = titles_and_tags['title'].astype('str').map(lambda x : len(x.split(" ")))
tags_length = titles_and_tags['tags'].astype('str').map(lambda x : len(x.split("|")))
df['title_length'] = titles_length
df['tags_length'] = tags_length

df

```

```python
df_select_id = df[["video_id"]]

df_select_binary = df[["description_contains_facebook","description_contains_instagram","title_contains_num","title_contains_special","title_contains_brackets","has_description"]]

# df_select_binary = df_select_binary.replace(np.nan, 'False', regex=True)

df_select_binary = df_select_binary.replace(np.nan, 0)
df_select_binary_ = df_select_binary.astype(int)

df_select_binary_
```

```python
df_select_string = df[["sub_date","channel frequency","category_of_title_length","tags_length_category"]]

df_select_num = df[["title_length","tags_length","daily_likes","daily_dislikes","daily_comment_count","stop_words_count_title","stop_words_count_tags","hour"]]

df_select_output = df[["daily_view_classification","daily_views"]]


x = df_select_num.values
# Create a minimum and maximum processor object
min_max_scaler = preprocessing.MinMaxScaler()

# Create an object to transform the data to fit minmax processor
x_scaled = min_max_scaler.fit_transform(x)

df_select_num_normalized = pd.DataFrame(x_scaled)
# Run the normalizer on the dataframe
# df_select_num_normalized[["title_length_normal","tags_length_normal","daily_likes_normal","daily_dislikes_normal","daily_comment_count_normal"]] = pd.DataFrame(x_scaled)

# df_select_number_normalized = df_select_num_normalized.iloc[:,5:10]

df_select_num_normalized[["title_length_normal","tags_length_normal","daily_likes_normal","daily_dislikes_normal","daily_comment_count_normal","stop_words_count_title_normal","stop_words_count_tags_normal","hour_normal"]] = pd.DataFrame(x_scaled)

df_select_number_normalized = df_select_num_normalized.iloc[:,8:16]

df_new = df_select_id.join(df_select_string)

df_new = df_new.join(df_select_binary_)

df_new = df_new.join(df_select_number_normalized)

df_new = df_new.join(df_select_output)

df_new.to_csv("normalize_input424.csv", sep=',')
print("done")
```
