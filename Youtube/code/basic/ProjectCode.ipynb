```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('USvideos.csv')
```

```python
df
```

```python
channel = df['channel_title'] #所有channel title
```

```python
channel_counts = df['channel_title'].value_counts() # 每个channel出现的次数
```

```python
channel_counts
```

```python
channel_counts.size/3 #所有channel平分3个等份
```

```python
fig = plt.figure()
plt.hist(channel_counts, bins = 20)
plt.title('Video Posting Count')
plt.xlabel('Channel Posting Frequency')
plt.show()
fig.savefig('Channel Video Posting Count.jpg',bbox_inches='tight')
```

```python
i = 1
while (channel_counts[channel_counts <= i].size)<= (channel_counts.size/3):
    i = i+1
    
i
#算出low frequency需要post多少video
```

```python
Low = channel_counts[channel_counts <= 5].size # Approxiamte one video in 43 days, f = 0.02
```

```python
Low #low frequency 的 channel数量
```

```python
i = 5
while (channel_counts[channel_counts <= i].size -Low) <= (channel_counts.size/3):
    i = i+1
    
i
#算出moderate frequency需要post多少video
```

```python
Moderate = channel_counts[channel_counts <= 14].size - Low  # Approximate one video in 15 days, f = 0.6
```

```python
Moderate #moderate frequency 的 channel数量
```

```python
High = channel_counts[channel_counts > 14].size 
```

```python
High #high frequency 的 channel数量
```

```python
channel_freq = []
for item in channel:
    if item in channel_counts[channel_counts <= 5]:
        channel_freq.append('low')
    elif item in channel_counts[channel_counts > 14]:
        channel_freq.append('high')
    else:
        channel_freq.append('moderate')
    
df.insert(16,'channel frequency',channel_freq) #加入新的column
df
```

```python
fig = plt.figure()
frequency = ('Low', 'Moderate', 'High')
y_pos = np.arange(len(frequency))
performance = [Low,Moderate,High]
plt.bar(y_pos, performance, align='center', alpha=0.5)
plt.xticks(y_pos, frequency)
plt.ylabel('Count')
plt.xlabel('Frequency')
plt.title('Channel Posting Frequency Count')
 
plt.show()
fig.savefig('Video Posting Bar Chart.jpg',bbox_inches='tight')
```
